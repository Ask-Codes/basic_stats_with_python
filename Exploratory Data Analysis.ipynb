# Exploratory Data Analysis

Import Pandas and Numpy to create a dataset of 1000 students' scores in four major subjects in public and private secondary schools.

```python
import pandas as pd
import numpy as np
```

```python
public = pd.DataFrame({'id1':np.arange(1,1001,1), 'English1':np.random.randint(40,90,1000),
                       'Maths1': np.random.randint(45,85,1000),
                       'Biology1': np.random.randint(40,80,1000),
                       'Physics1': np.random.randint(45,90,1000)})
```

```python
private = pd.DataFrame({'id2':np.arange(1,1001,1), 'English2':np.random.randint(40,90,1000),
                       'Maths2': np.random.randint(45,85,1000),
                       'Biology2': np.random.randint(40,80,1000),
                       'Physics2': np.random.randint(45,90,1000)})
```

```python
print(public.head()) # shows the first N rows in the data (by defualt, N=5)
```

```python
print(private.head())# shows the first N rows in the data (by defualt, N=5)
```

```python
student_scores = public.join(private)
```

# Correlation between public and private schools performance

```python
correlation = student_scores.corr(method="pearson")
print(correlation)
```

```python
import seaborn as sns #to harmonize the correlation
```

```python
sns.heatmap(correlation,xticklabels=correlation.columns, yticklabels=correlation.columns, annot = True)
```

```python
sns.pairplot(student_scores) #plots the bar and scatter plots that compares the relationship between the performance of studetn in public and private schools
```
